# Question 5: Data Consistency & Quality Checks

```python
# Change the datatype of the column 'user_id' to string as 'object'

df_customers['user_id'] = df_customers['user_id'].astype('str')
```

```python
# Displays information about the dataframe

df_customers.info()
```

```python
# Check for mixed data types

for col in df_customers.columns.tolist():
    weird = (df_customers[[col]].map(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_customers[weird]) > 0:
        print(col)
```

# The column 'First name' has mixed data types. Hence will set the datatype for the column to be string.

```python
# Addressing mixed data types

df_customers['First_Name'] = df_customers['First_Name'].astype('str')
```

# Question 6: Combining Customer data & prepared Instacart Data

```python
# Check the output of the imported dataframe

ords_prods.head()
```

```python
ords_prods.info()
```

```python
# Checking the datatype of the column 'user_id' in ords_prods

ords_prods['user_id'].dtype
```

```python
# Changing the datatype of the column 'user_id' in ords_prods dataframe to string

ords_prods['user_id'] = ords_prods['user_id'].astype('str')
```

```python
# Dropping the column _merge as it would not be relevant

ords_prods = ords_prods.drop(columns = ['_merge'])
```

```python
# Combining the customer dataset and the Instacart dataset using the variable 'user_id' as the key

orders_products_customers = ords_prods.merge(df_customers, on = ['user_id'], indicator = True)
```
