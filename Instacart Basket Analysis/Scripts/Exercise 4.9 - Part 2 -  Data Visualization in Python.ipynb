# Question 7: Checking if there is a connection between age and spending power(income)

```python
# Creating a scatterplot with the variables 'age' and 'income'

scatterplot_age_income = sns.scatterplot(x = 'Age', y = 'income', data = ords_prods)
```

# The scatterplot shows that the maximum income of all the age groups is between 0 and 200,000 dollars. Then the next highest income is between 200,000 dollars and 300,000 dollars which are from the 40+ age groups. Between 200,000 dollars and 400,000 dollars, the income values are distributed evenly across all age groups. And the incomes of 400,000 dollars to 600,000 dollars is mainly only from the age groups 40 and above.

```python
# Exporting the scatterplot

scatterplot_age_income.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'scatterplot_age_income.png'))
```
