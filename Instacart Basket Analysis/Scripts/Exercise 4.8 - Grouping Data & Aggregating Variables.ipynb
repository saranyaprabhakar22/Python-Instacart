```python
# Finding the aggregated mean of the 'order_number' column grouped by the 'department_id' column for the entire dataframe

df_ords_prods_merge.groupby('department_id').agg({'order_number' : ['mean']})
```

# Question 4: Creating a loyalty flag for existing customers

```python
# Creating a loyalty flag
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge.head(15)
```

```python
# Deriving columns with loc()
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check the output with the newly created 'loyalty_flag' column
df_ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(30)
```

# Question 5: Basic statistics of the product prices for each loyalty category

```python
# Finding the mean, maximum and minimum for each of the records in the loyalty column
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'max', 'min']})
```

# Question 6: Spending flag for each user

```python
# Creating a flag for each user
df_ords_prods_merge['average_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge.head(10)
```

```python
# Creating columns
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check the output
df_ords_prods_merge.head()
```

# Question 7: Determine frequent versus non-frequent customers

```python
# Creating a flag for each user
df_ords_prods_merge['median'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merge.head(10)
```

```python
# Creating columns
df_ords_prods_merge.loc[df_ords_prods_merge['median'] > 20, 'frequency_flag'] = 'Non-frequent Customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median'] > 10) & (df_ords_prods_merge['median'] <= 20), 'frequency_flag'] = 'Regular Customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
df_ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Looking for values above 100 in the prices column and assigning them a value
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
df_ords_prods_merge['prices'].max()
```
