# Create a user-defined function

```python
# Define a function
def price_label(row):

  if row['prices'] <= 5:
        return 'Low-range Product'
  elif (row['prices'] > 5) or (row['prices'] <= 15):
        return 'Mid-range Product'
  elif row['prices'] > 15:
        return 'High-range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Check the most expensive product
df['prices'].max()
```

# If statements with the loc() function

```python
# Create conditions with loc()
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Using the loc() function in the entire dataframe ords_prods_merged
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

# If-Statements with for loop

```python
# Display the frequency of the orders_days_of_week column
df_ords_prods_merge['orders_days_of_week'].value_counts(dropna = False)
```

```python
# Creating the for loop

result = []

for value in df_ords_prods_merge['orders_days_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
         result.append('Least busy')
    else: 
        result.append('Regularly busy')
```

```python
# Check the output
result
```

```python
# Combining with the ords_prods_merge dataframe
df_ords_prods_merge['busiest day'] = result
```

```python
df_ords_prods_merge['busiest day'].value_counts(dropna = False)
```

# Question 2: Changing the labels in the 'busiest day' column

```python
# Creating the for loop

result = []

for value in df_ords_prods_merge['orders_days_of_week']:
    if value == 0 or value == 1:
        result.append('Busiest days')
    elif value == 3 or value == 4: 
         result.append('Slowest days')
    else: 
        result.append('Regular days')
```

# Question 4: Identify the busiest hours of the day

```python
# Display the frequency of the order_hour_of_day column
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Creating the for loop

result = []

for value in df_ords_prods_merge['order_hour_of_day']:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
        result.append('Most orders')
    elif value == 17 or value == 8 or value == 18 or value == 19 or value == 20 or value == 7 or value == 21 or value == 22: 
         result.append('Average orders')
    else: 
        result.append('Fewest orders')
```

```python
# Check the output
result
```

```python
# Combining with the dataframe
df_ords_prods_merge['busiest_period_of_day'] = result
```
